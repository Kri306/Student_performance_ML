```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Load your CSV
data = pd.read_csv("C:\\python\\Student_performance_ML\\Student_performance_data _.csv")

# Features and target
X = data[['StudyTimeWeekly']]
y = data['GPA']

# Train model on all data
model = LinearRegression()
model.fit(X, y)

# Predict on the same data
y_pred = model.predict(X)

# Regression metrics
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print("MAE:", round(mae, 2))
print("MSE:", round(mse, 2))
print("RMSE:", round(rmse, 2))
print("R^2 Score:", round(r2, 4))

# ----------------------------
# Plot histogram of residuals
# ----------------------------

residuals = y - y_pred

plt.figure(figsize=(8, 5))
plt.hist(residuals, bins=20, edgecolor='black', color='skyblue')
plt.title('Histogram of Residuals')
plt.xlabel('Residual (Actual GPA - Predicted GPA)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
